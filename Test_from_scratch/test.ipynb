```python
# read all .wav data from root = './data/unlabeled_test/' with torch audio and calculate the length in second of each audio
def read_audio(root):
    audio_list = []
    audio_length = []
    for filename in os.listdir(root):
        if filename.endswith(".WAV"):
            audio_list.append(os.path.join(root, filename))
            waveform, sample_rate = torchaudio.load(os.path.join(root, filename))
            audio_length.append(waveform.shape[1]/sample_rate)
    return audio_list, audio_length

root = './data/unlabeled_test/'
audio_list, audio_length = read_audio(root)

#make df 
df = pd.DataFrame({'audio_list':audio_list, 'audio_length':audio_length})
```

```python
def calculate_mel_spec_size(sample_rate, desired_length_in_seconds, n_fft, hop_length, n_mels):
    number_of_samples = sample_rate * desired_length_in_seconds
    number_of_frames = (number_of_samples // hop_length) + 1  # Approximation
    return n_mels, number_of_frames

# Example usage
sample_rate = 32000
desired_length_in_seconds = 11  # Example length
n_fft = 2048
hop_length = 512
n_mels = 128
mel_spec_height, mel_spec_width = calculate_mel_spec_size(sample_rate, desired_length_in_seconds, n_fft, hop_length, n_mels)
print(f"Mel Spectrogram Size: {mel_spec_height} x {mel_spec_width}")
```
