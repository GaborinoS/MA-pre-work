```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
f1_SA = pd.read_pickle("results_standalone/SA_ADSMI_RS96_F1_scores/macro_avg_f1_scores.pkl") 
f1_FT = pd.read_pickle("finetune_results/FT_ADSMI_2024-05-07-18-04-53_Mavg/f1_scores.pkl")

# Ensure the data is in the right format for cummax
f1_SA = np.array(f1_SA) 
f1_FT = np.array(f1_FT) 

# Plotting the F1 Scores
plt.figure(figsize=(10, 7))
plt.plot(f1_SA[0:44], label='Standalone')
plt.plot(f1_FT[0:44], label='Fine-Tuned')
plt.title("macro avg F1 Scores")
plt.legend()
plt.show()

# Make a cumulative maximum of the F1 scores and plot it
f1_SA_cummax = np.maximum.accumulate(f1_SA)
f1_FT_cummax = np.maximum.accumulate(f1_FT)


plt.figure(figsize=(10, 7))
plt.plot(f1_SA_cummax[0:44], label='Standalone')
plt.plot(f1_FT_cummax[0:44], label='Fine-Tuned')
plt.title("Cumulative Max macro avg F1 Scores")
plt.legend()
plt.show()
```
