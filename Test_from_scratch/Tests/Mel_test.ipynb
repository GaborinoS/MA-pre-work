```python
import torchaudio
import torchaudio.transforms as T
import matplotlib.pyplot as plt

# Load audio data
waveform, sample_rate = torchaudio.load("./Test_from_scratch/data/ESC50/ESC-50-master/audio/1-137-A-32.wav")

# Convert to spectogram
transform = T.MelSpectrogram(sample_rate=51200, n_mels=128, n_fft=1024, hop_length=512,)
spectrogram = transform(waveform)

#plot spectrogram librosa
import librosa.display
import numpy as np

# Convert to numpy array
spectrogram = spectrogram.numpy()[0]

# Convert to decibels
spectrogram = librosa.power_to_db(spectrogram, ref=np.max)

# Plot
librosa.display.specshow(spectrogram, y_axis='mel', fmax=8000, x_axis='time')
plt.colorbar(format='%+2.0f dB')
plt.title('Mel spectrogram')
plt.tight_layout()
plt.show()

```
